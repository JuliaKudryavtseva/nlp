```python
def go_deeper(synset):
    synsets = []
    hyponyms = synset.hyponyms()
    if hyponyms == []:
        return synsets
    synsets.extend(hyponyms)
    for hyponym in hyponyms:
        synsets.extend(go_deeper(hyponym))
    return synsets  
```

```python
class PairsDataset(torch.utils.data.Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __getitem__(self, idx):
        assert idx <= len(self.x['input_ids']), (idx, len(self.x['input_ids']))
        item = {key: val[idx] for key, val in self.x.items()}
        item['decoder_attention_mask'] = self.y['attention_mask'][idx]
        item['labels'] = self.y['input_ids'][idx]
        return item
    
    @property
    def n(self):
        return len(self.x['input_ids'])

    def __len__(self):
        return self.n
```

```python
from typing import List, Dict, Union

class DataCollatorWithPadding:
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        batch = self.tokenizer.pad(
            features,
            padding=True,
        )
        ybatch = self.tokenizer.pad(
            {'input_ids': batch['labels'], 'attention_mask': batch['decoder_attention_mask']},
            padding=True,
        ) 
        batch['labels'] = ybatch['input_ids']
        batch['decoder_attention_mask'] = ybatch['attention_mask']
        
        return {k: torch.tensor(v) for k, v in batch.items()}
```

```python
def evaluate_model(model, test_dataloader):
    num = 0
    den = 0

    for batch in test_dataloader:
        with torch.no_grad():
            loss = model(**{k: v.to(model.device) for k, v in batch.items()}).loss
            num += len(batch) * loss.item()
            den += len(batch)
    val_loss = num / den
    return val_loss
```

```python
df_input = []
df_output = []
test_synsets_to_exclude = [j for i in parents_with_children.values() for j in i] + [i for i in parents_with_children]


for synset in wn.all_synsets("n"):
    if synset not in test_synsets_to_exclude:
        hypernyms = ", ".join([i.name().replace("_", " ") for i in synset.hypernyms()])
        hyperhypernyms = ", ".join([j.name().replace("_", " ") for i in synset.hypernyms() for j in i.hypernyms()])
        hyponyms = [j.name().replace("_", " ") for i in synset.hyponyms() for j in i.lemmas()]
        synset_name = synset.name().replace("_", " ")
        if len(hyponyms) > 0:
            for ind in range(0, len([j.name() for i in synset.hyponyms() for j in i.lemmas()]), 10):
                cur_hyponyms = ", ".join(hyponyms[ind:ind+10])
                input_text = f"hyperhypernyms: {hyperhypernyms} | hypernyms: {hypernyms} | synset: {synset_name} | hyponyms:"
                output_text = cur_hyponyms
                df_input.append(input_text)
                df_output.append(output_text)
```

```python
df_test_input = []
df_test_output = []

for synset, lemmas in parents_with_lemmas.items():
    hypernyms = ", ".join([i.name().replace("_", " ") for i in synset.hypernyms()])
    hyperhypernyms = ", ".join([j.name().replace("_", " ") for i in synset.hypernyms() for j in i.hypernyms()])
    hyponyms = [j.name().replace("_", " ") for i in synset.hyponyms() for j in i.lemmas()]
    synset_name = synset.name().replace("_", " ")
    cur_hyponyms = ", ".join(lemmas)
    input_text = f"hyperhypernyms: {hyperhypernyms} | hypernyms: {hypernyms} | synset: {synset_name} | hyponyms:"
    output_text = cur_hyponyms
    df_test_input.append(input_text)
    df_test_output.append(output_text)
```

```python
final_scores = []
prec_1 = []
prec_2 = []
prec_5 = []
prec_10 = []


with open("example_t5_large_correct-notrain2-50-norepeat2_max_len_2048_max_50_except_wn_numbeams_5_deversity_1.0", "w") as w:
    for i_2, j_2 in zip(test_df['toxic_comment'].tolist(), test_df['neutral_comment'].tolist()):
        #w.write(i_2 + "\n")
        input_ids = tokenizer.encode(i_2, return_tensors="pt")
        output_batch = model.generate(input_ids.cuda(), no_repeat_ngram_size=2, max_new_tokens=2048, 
                                      num_return_sequences=50, num_beams=50, early_stopping=True, 
                                      num_beam_groups=5, 
                                      diversity_penalty=1.0)
        decoded_list = []
        for outputs in output_batch:
            decoded = tokenizer.decode(outputs, skip_special_tokens=True).split(", ")
            decoded_list.extend(decoded)

        predicted_answer = set(decoded_list)
        sorted_predicted_answer = [i[0] for i in Counter(decoded_list).most_common()]
        
        true_answers = set(j_2.split(", "))
        
        scores = len(true_answers.intersection(predicted_answer))
        pr_1 = len(true_answers.intersection(set(sorted_predicted_answer[:1])))/1
        pr_2 = len(true_answers.intersection(set(sorted_predicted_answer[:2])))/2
        pr_5 = len(true_answers.intersection(set(sorted_predicted_answer[:5])))/5
        pr_10 = len(true_answers.intersection(set(sorted_predicted_answer[:10])))/10
        
        final_scores.append(scores/len(predicted_answer))
        prec_1.append(pr_1)
        prec_2.append(pr_2)
        prec_5.append(pr_5)
        prec_10.append(pr_10)
        
        
        
        w.write(i_2+"\n predicted: \n")
        w.write(", ".join(sorted_predicted_answer) + "\n")
        w.write(" true: \n"+ j_2 + "\n")
        w.write(f"pr@1:{pr_1}, pr@2:{pr_2}, pr@5:{pr_5}, pr@10:{pr_10}, pr_total:{scores/len(predicted_answer)}\n")
        w.write("-"*10)
        w.write("\n")
    w.write(f"total pr@1: {sum(prec_1)/len(prec_1)}\n")
    w.write(f"total pr@2: {sum(prec_2)/len(prec_2)}\n")
    w.write(f"total pr@5: {sum(prec_5)/len(prec_5)}\n")
    w.write(f"total pr@10: {sum(prec_10)/len(prec_10)}\n")
    w.write(f"total pr: {sum(final_scores)/len(final_scores)}\n")
```

```python
def _metrics():
    fin_sc = f'aggreagted final_scores: {sum(final_scores)/len(final_scores)}'
    
    prec_1_sc = f'aggreagted prec_1: {sum(prec_1)/len(prec_1)}'
    
    prec_2_sc = f'aggreagted prec_2: {sum(prec_2)/len(prec_2)}'

    prec_5_sc = f'aggreagted prec_1: {sum(prec_1)/len(prec_1)}'

    prec_10_sc = f'aggreagted prec_1: {sum(prec_1)/len(prec_1)}'
    return fin_sc + '\n' + prec_1_sc  + '\n' + prec_2_sc  + '\n' +  prec_5_sc + '\n' + prec_10_sc
```
